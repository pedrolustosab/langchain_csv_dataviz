```python
import os
from dotenv import load_dotenv
import pandas as pd
import plotly.express as px
from langchain.agents import initialize_agent, Tool
from langchain.llms import OpenAI
from langchain.agents import AgentType
from langchain.prompts import PromptTemplate
from langchain_experimental.tools import PythonAstREPLTool
from langchain.chat_models import ChatOpenAI
from langchain_experimental.agents import create_pandas_dataframe_agent
from tabulate import tabulate
```

```python
# Load environment variables from .env
load_dotenv()

# Load the CSV file into a pandas DataFrame
df = pd.read_csv("basededados_aulat7.csv")
```

```python
# Initialize the LLM and create the pandas DataFrame agent
llm = OpenAI(model="gpt-4o", temperature=0)
agent_executor = create_pandas_dataframe_agent(llm, df, agent_type="openai-tools", verbose=True, allow_dangerous_code=True)
```

```python
# Define the prompt template
prompt_template = """
Você é um especialista em manipulação de dados com Pandas DataFrames. Abaixo, você receberá uma descrição em linguagem natural de um filtro que deve ser aplicado a um DataFrame chamado 'df'. 

Sua tarefa é gerar o código Pandas correspondente para filtrar o DataFrame conforme a descrição fornecida. Certifique-se de que o código:
1. Use a função de filtragem correta do Pandas (como `loc`, `query`, ou `boolean indexing`).
2. Considere a estrutura do DataFrame e os tipos de dados possíveis.
3. Retorne um DataFrame filtrado baseado na descrição fornecida.

Descrição do filtro: {description}

Forneça o código Pandas necessário para obter o DataFrame filtrado. 

Aqui estão alguns exemplos para ilustrar possíveis descrições e seus filtros correspondentes:

- Descrição: "Filtre todas as linhas onde a coluna 'idade' é maior que 30."
  Código: `df_filtered = df[df['idade'] > 30]`

- Descrição: "Selecione as linhas onde a coluna 'categoria' é igual a 'A' e a coluna 'valor' é menor que 100."
  Código: `df_filtered = df[(df['categoria'] == 'A') & (df['valor'] < 100)]`

Ajuste o código conforme necessário para corresponder à descrição fornecida.
"""

df_prompt = PromptTemplate(template=prompt_template, input_variables=["description"])
```

```python
# Define a description for filtering
description = "Me de a quant total por vendedor."

# Generate the Pandas code using the agent executor
code = agent_executor.run(df_prompt.format(description=description))

# Print the generated code
print("Generated Pandas code:\n", code)

# Execute the generated code to filter the DataFrame
# Here, we'll assume that the code is safe and executable
# In practice, you should handle code execution with care
exec(code)

# Display the filtered DataFrame
print("Filtered DataFrame:")
print(df_filtered.head())
```
